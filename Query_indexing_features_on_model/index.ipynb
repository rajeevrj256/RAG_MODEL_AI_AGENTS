```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['GOOGLE_GEMINI_AI_API_KEY']=os.getenv("GOOGLE_GEMINI_AI_API_KEY")
```

```python
# imports
from llama_index.embeddings.gemini import GeminiEmbedding
```

```python
embedding=GeminiEmbedding(model="models/embedding-001",google_api_key="GOOGLE_GEMINI_AI_API_KEY")
```

```python
embedding
```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader
documents=SimpleDirectoryReader("data").load_data()
```

```python
index=VectorStoreIndex.from_documents(documents,
                                      embed_model=embedding, 
                                      show_progress=True)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("Name of candidate")
```
